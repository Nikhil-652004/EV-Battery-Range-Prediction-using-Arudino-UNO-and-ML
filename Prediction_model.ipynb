Random forest

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Load the synthetic dataset
df = pd.read_csv('/content/drive/MyDrive/IoT_Project/synthetic_ev_dataset_accurate.csv')

# Features: Current, Voltage, Speed, Temperature, SOC
X = df[['Current', 'Voltage', 'Speed', 'Temperature', 'SOC']]

# Target: Remaining Range
y_remaining_range = (df['SOC'] * df['Range']) / 100

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_remaining_range, test_size=0.2, random_state=42)

# Create a Random Forest regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error for Remaining Range (Random Forest): {mse}')
```

```python
from joblib import dump

# Save the trained Random Forest model
model_filename = 'random_forest_model.joblib'
dump(model, model_filename)
```
